### Removing the skewness using yeojohnson transformation metod

```python
from scipy.stats import yeojohnson
from sklearn.preprocessing import PowerTransformer

# yeo johnson transformation on highly skewed column

df['R']=yeojohnson(df['R'], 0)
df['E']=yeojohnson(df['E'], 0)
df['CG']=yeojohnson(df['CG'], 0)
df['H']=yeojohnson(df['H'], 0)
df['SV']=yeojohnson(df['SV'], 0)
df['SHO']=yeojohnson(df['SHO'], 0)
df['HR']=yeojohnson(df['HR'], 0)
```

```python
# Validating the skewness
sk=df.skew()
sk[np.abs(sk)>0.5].all()
```

As we see the above showing true so we can say that skewness is removed within range(-0.5-0.5)

### Data scaling

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

## Split Data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=1352)
```

## Model instansiation and trainig
### 1. Linear regression

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```

### Adjusted r2 square

```python
# For training data
regression.score(x_train,y_train)*100
```

```python
# For testing data
regression.score(x_test,y_test)*100
```

```python
y_pred=regression.predict(x_test)
```

```python
y_pred
```

### Lets plot and visualize

```python
plt.scatter(y_test,y_pred)
plt.xlabel('actual winning percentage')
plt.ylabel('predicted winning percentage')
plt.title('Actual vs model predicted')
plt.show()
```

### Model evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

### 2. Decision tree regression

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
dt=DecisionTreeRegressor(random_state=0)
dt.fit(x_train,y_train)
```

```python
preddt=dt.predict(x_test)
preddt
```

```python
# plot the graph
plt.scatter(y_test,preddt)
plt.xlabel('actual winning percentage')
plt.ylabel('predicted winning percentage')
plt.title('Actual vs model predicted')
plt.show()
```

```python
# model evaluation
mean_absolute_error(y_test,preddt)
```

```python
mean_squared_error(y_test,preddt)
```

### 3.Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
predrf
```

```python
# plot and visualize
plt.scatter(y_test,predrf)
plt.xlabel('actual winning percentage')
plt.ylabel('predicted winning percentage')
plt.title('Actual vs model predicted')
plt.show()
```

```python
# model evaluation
mean_absolute_error(y_test,predrf)
```

```python
mean_squared_error(y_test,predrf)
```

### Cross validation for overfitting

```python
from sklearn.model_selection import cross_val_score
scr=cross_val_score(regression,x_scaled,y,cv=5)
print(scr.mean())
```
